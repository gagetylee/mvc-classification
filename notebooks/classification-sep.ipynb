```python
# Define global variables
import sys
import os

module_path = os.path.abspath(os.path.join('..', 'src'))
if module_path not in sys.path:
    sys.path.append(module_path)

S_FREQ = 2048
```

```python
# Load metadata
import os
import pandas as pd
import numpy as np
from data.make_dataset import load_metadata, load_signals

metadata = load_metadata('../dataset')
metadata
```

```python
# Load signals
signals = load_signals(metadata, nrows=10)
signals
```

```python
# Channel selection
from preprocessing.channel_selection_np import extract_channels

selected_channels = extract_channels(signals, 10)
```

```python
# Preprocess signal data
from preprocessing.preprocessing import preprocess_signal

processed_df = preprocess_signal(selected_channels, S_FREQ)
processed_df
```

```python
# create feature table
from features.build_features import feature_table

fft = feature_table(processed_df, metadata, S_FREQ)
fft
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Split and scale data
X = fft.drop(['signal', 'channel', '%mvc'], axis=1)
y = fft['%mvc']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# train model
from model.train_model import train_model

model = train_model(X_train_scaled, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Evaluate model
y_pred = model.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f'Accuracy: {accuracy}')
print(report)

conf_matrix = confusion_matrix(y_test, y_pred)
sns.set_context('talk')

ax = sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['10%','30%','50%'], yticklabels=['10%','30%','50%'])
ax.xaxis.tick_top()
ax.set_xlabel('Predicted', labelpad=10)
ax.set_ylabel('Actual', labelpad=10)
ax.xaxis.set_label_position('top')

plt.show()
```
