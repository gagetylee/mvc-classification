```python
# Define global variables
import sys
import os

module_path = os.path.abspath(os.path.join('..', 'src'))
if module_path not in sys.path:
    sys.path.append(module_path)

S_FREQ = 2048
S_DURATION = 10
```

```python
# Load metadata
import os
import pandas as pd
import numpy as np
from data.make_dataset import load_metadata, load_signals

metadata = load_metadata('../dataset')
metadata
```

```python
# Load signals
signals = load_signals(metadata, nrows=144)
signals
```

```python
# Channel selection
from preprocessing.channel_selection_np import extract_channels

selected_channels = extract_channels(signals, 16)
```

```python
# Preprocess signal data
from preprocessing.preprocessing import butter_bandpass_filter, rectify_and_smooth

def preprocess_signal(df, fs):
    lowcut = 10.0
    highcut = 500.0
    order = 4

    copy = df.copy()
    # Band-pass filter
    processed_df = butter_bandpass_filter(copy, lowcut, highcut, fs, order)
    # Rectification
    processed_df = rectify_and_smooth(processed_df)
    
    return processed_df

processed_df = preprocess_signal(selected_channels, S_FREQ)
processed_df
```

```python
import matplotlib.pyplot as plt

time_a = np.linspace(0, S_DURATION, int(S_FREQ * S_DURATION))
emg_a = signals.loc[0][6]

time_b = np.linspace(0, S_DURATION, int(S_FREQ * S_DURATION))
emg_b = processed_df.loc[0][6]


plt.figure(figsize=(20, 5))

plt.subplot(1, 2, 1)
plt.plot(time_a, emg_a, linewidth=1)
plt.xlim([0, 5])
plt.title('Single-channel EMG')
plt.xlabel('time (s)')
plt.ylabel('Âµv')

plt.subplot(1, 2, 2)
plt.plot(time_b, emg_b, linewidth=2)
plt.xlim([0, 5])
plt.title('Single-channel EMG (Preprocessed)')
plt.xlabel('time (s)')
plt.ylabel('Âµv')

plt.show()
print(int(S_FREQ * S_DURATION))
```

```python
# create feature table
from features.build_features import feature_table

fft = feature_table(processed_df, metadata, S_FREQ)
fft
```

```python
from sklearn.model_selection import train_test_split

mvc_mapping = fft.reset_index().groupby('signal')['%mvc'].first()

# Extract unique signals
unique_signals = fft.index.get_level_values('signal').unique()
stratify_arr = mvc_mapping.loc[unique_signals].values

# Split signals into training and testing sets
train_signals, test_signals = train_test_split(unique_signals, test_size=0.2, random_state=45, stratify=stratify_arr)

# Create training and testing DataFrames
train_df = fft.loc[train_signals]
test_df = fft.loc[test_signals]

features = ['rms', 'mav', 'wl', 'mean_freq']

X_train = train_df[features]
y_train = train_df['%mvc']
X_test = test_df[features]
y_test = test_df['%mvc'].groupby(level='signal').first()

# X_train = train_df.drop(['%mvc', 'pca_coeff'], axis=1)
# y_train = train_df['%mvc']
# X_test = test_df.drop(['%mvc', 'pca_coeff'], axis=1)
# # y_test = test_df['%mvc']
# y_test = test_df['%mvc'].groupby(level='signal').first()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
cols_to_scale = ['rms', 'mav', 'wl', 'mean_freq']

X_train_scaled = scaler.fit_transform(X_train[cols_to_scale])
X_test_scaled = scaler.transform(X_test[cols_to_scale])

X_train[cols_to_scale] = X_train_scaled
X_test[cols_to_scale] = X_test_scaled
```

```python
# # train model
from model.train_model import train_model

model = train_model(X_train, y_train)
```

```python
# predictions
predictions = {}
for (signal, channel), data in X_test.groupby(level=['signal', 'channel']):
    predictions[(signal, channel)] = model.predict(data)[0]
    
data = [(signal, channel, pred) for (signal, channel), pred in predictions.items()]
predictions_df = pd.DataFrame(data, columns=['signal', 'channel', 'prediction'])
predictions_df = predictions_df.pivot(index='signal', columns='channel', values='prediction')
predictions_df

final_predictions = predictions_df.mode(axis=1)[0]
```

```python
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Evaluate model


accuracy = accuracy_score(y_test, final_predictions)
report = classification_report(y_test, final_predictions)

print(f'Accuracy: {accuracy}')
print(report)

conf_matrix = confusion_matrix(y_test, final_predictions)
sns.set_context('talk')

ax = sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['10%','30%','50%'], yticklabels=['10%','30%','50%'])
ax.xaxis.tick_top()
ax.set_xlabel('Predicted', labelpad=10)
ax.set_ylabel('Actual', labelpad=10)
ax.xaxis.set_label_position('top')

plt.show()
```
